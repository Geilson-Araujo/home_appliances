```python
import os
from urllib.request import urlretrieve
from urllib.error import HTTPError, URLError

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import requests
from bs4 import BeautifulSoup

from tqdm.notebook import tqdm

pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)
```

```python
def appliance_scrapping(url: str, path: str):

    # Define headers
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Brave Chrome/84.0.4147.89 Safari/537.36'}    
     
    # Send a GET request
    r = requests.get(url, headers=headers)    

    # Define econding
    r.encoding = 'utf-8' 

    # Parse the html
    html = r.text
    soup = BeautifulSoup(html, 'html.parser')

    # Compute the pages number
    pages_num = soup.find('div', {'id': 'quantidadePaginas'}).get_text().split()[-1]
    
    for num in tqdm(range(int(pages_num))):
                
        # Select de URL by page number
        r = requests.get(url + '?p=' + str(num + 1), headers=headers)
        r.encoding = 'utf-8' # ISO-8859-1
            
        # try:
        # Parse the html
        html = r.text
        soup = BeautifulSoup(html, 'html.parser')
        # except HTTPErr:
        # print('{} not found'.format(url))
        # continue
    
        # Download the images
        for item in soup.find_all('img', {'class': 'center-block lazy'}):
            
            # Define the URL
            url_ = 'http:' + item.get('data-original')
        
            # Compose the image name
            prefix = item.get('title').lower().replace(' ', '_')
            suffix = item.get('data-original').split('/')[-1]
            img_name = prefix + '_' + suffix
        
            # Define the file location
            filename = os.path.join(path, img_name)
            
            try:
                urlretrieve(url_, filename)
            except HTTPError as e:
                print(e.status, e.reason)
                continue
            except URLError as e:
                print(e.reason)
                continue
```
